```python
import pandas as pd
from pandas import DataFrame

import numpy as np
```

파일 불러오기

```python
new = pd.read_csv('data-house-warming-with-all-proudct-reviews.csv', sep='\t', encoding="UTF-16")
```

파일 전처리 끝 
전체 데이터 중 분석에 사용할 데이터만 남기기

```python
new_data_review_avg = new_2.drop(['review_text', 'review_count', 'production_info_is_purchased', 'review_user_id', 'house_residence',
                                  'house_area', 'house_expertise', 'house_color_list', 'house_style_list', 'house_family_list', 
                                  'review_star_avg'], axis=1)
```

```python
new_data_review_avg.head(5)
```

데이터들이 house 정보 기준으로 작성되어 중복된 데이터가 많아 분석 처리전 삭제

```python
new_data_review_avg = new_data_review_avg.drop_duplicates()
```

```python
new_data_review_avg
```

제품별 별점과 해당 제품의 스크랩횟수, 사람들이 확인한 횟수와 어떤 관계가 있는지 상관 계수로 확인

```python
review_corr = new_data_review_avg.loc[:, ['review_average', 'scrap_count', 'view_count']]
```

```python
review_corr.corr(method='pearson')
```

해당 제품이 사용된 인테리어 게시물의 like 수 와 해당 제품이 사용된 인테리어를 스크랩 한 수, 인테리어를 사람들이 확인한 횟수가 
어떤 관계가 있는지 상관 계수로 확인

```python
house_corr = new_data_review_avg.loc[:, ['house_like_count', 'house_scrap_count', 'house_view_count']]
```

```python
house_corr.corr(method='pearson')
```

인테리어 관련 score들을 하나로 정리 (각 컬럼 값 * 상관 계수 해서 더함)

```python
house_corr_col = DataFrame(house_corr['house_like_count'] + (house_corr['house_scrap_count']*0.980197) 
                           + (house_corr['house_view_count']*0.909834))
```

```python
new_data_review_avg_final = new_data_review_avg.loc[:, ['id', 'name', 'review_average']]
```

인테리어 score 값이 각 제품별 star 점수와 차이가 너무 커 log 사용해서 값을 맞춤

```python
new_data_review_avg_final['house_score'] = np.log10(house_corr_col[0])
```

```python
new_data_review_avg_final
```

제품의 별점과 인테리어 점수의 상관 계수를 구해줌

```python
(new_data_review_avg_final.loc[:, ['review_average','house_score']]).corr(method='pearson')
```

점수를 총합한 신규 테이블 생성

```python
review_avg_score = DataFrame(new_data_review_avg_final['review_average'] + (new_data_review_avg_final['house_score']*0.132855))
```

```python
review_avg_score.head(5)
```

```python
new_data_review_avg_final['score'] = review_avg_score[0]
```

```python
new_data_review_avg_final
```

점수를 기준으로 내림차순으로 데이터 정리해서 상위 제품부터 볼 수 있도록 함

```python
new_data_review_avg_final.sort_values(by='score', ascending=False)
```
