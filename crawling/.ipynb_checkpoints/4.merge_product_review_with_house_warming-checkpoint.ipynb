```python
import pandas as pd
import os
import numpy as np

df_warming = pd.read_csv('./input/data-house-warming.csv', sep='\t', encoding="UTF-16")
df_data = pd.read_csv('./input/data-product-review-v2.csv', sep='\t', encoding="UTF-16")
```

```python
df_data.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)
```

```python
df_warming.drop(['Unnamed: 0'], axis=1)
```

```python
merged_df = pd.DataFrame()
```

```python
for row in df_warming.itertuples(name='HouseWarming'):
    product_ids = row.purchased_product_ids.replace('[', '').replace(']', '').split(', ')
    if product_ids is None or len(product_ids) <= 0:
        continue

    for product_id in product_ids:
        found_df_row = df_data[lambda x: x['id'] == int(product_id)]
        #print('prodcut_id:', product_id, ', found_df:', found_df_row)
        
        if found_df_row.empty is True:
            continue
        new_df_row = found_df_row
        prefix = 'house_'
        
        new_df_row[prefix + 'residence'] = row.residence
        new_df_row[prefix + 'area'] = row.area
        new_df_row[prefix + 'region'] = row.region
        new_df_row[prefix + 'expertise'] = row.expertise
        new_df_row[prefix + 'color_list'] = row.color_list
        new_df_row[prefix + 'style_list'] = row.style_list
        new_df_row[prefix + 'constructions'] = row.constructions
        new_df_row[prefix + 'family_list'] = row.family_list.replace('[', '').replace(']', '').replace("'", '')
        new_df_row[prefix + 'color_list'] = row.color_list
        new_df_row[prefix + 'like_count'] = row.like_count
        new_df_row[prefix + 'reply_count'] = row.reply_count
        new_df_row[prefix + 'scrap_count'] = row.scrap_count
        new_df_row[prefix + 'view_count'] = row.view_count
        new_df_row[prefix + 'share_count'] = row.share_count
        
        merged_df = pd.concat([merged_df, new_df_row])
```

```python
new_df = pd.concat([merged_df, df_data])
```
